```python
import os
import pandas as pd

# Define directories
online_gaming_dir = r'C:\Users\Online Gaming'
video_streaming_dir = r'C:\Users\\Video Streaming'

# Define Video Streaming subfolders and associated ISPs
video_folders_orgs = {
    "Netflix": "Netflix",
    "PrimeVideo": "Amazon",
    "Crave": "Fastly"
}

# Function to filter the Online Gaming CSV files
def filter_online_gaming(file_path):
    df = pd.read_csv(file_path)
    # Filter rows where ISP contains "Riot Games" (partial match, ignoring case)
    filtered_df = df[df['Org'].str.contains('Riot Games', case=False, na=False)]
    # Also keep rows where Source IP is in the Destination IP (for Riot Games)
    riot_ip_rows = df[df['Destination'].isin(filtered_df['Source'])]
    final_df = pd.concat([filtered_df, riot_ip_rows]).drop_duplicates()
    # Save the filtered CSV
    save_path = file_path.replace('.csv', '_filtered.csv')
    final_df.to_csv(save_path, index=False)
    print(f"Filtered Online Gaming file saved to {save_path}")

# Function to filter the Video Streaming CSV files based on folder name
def filter_video_streaming(file_path, isp):
    df = pd.read_csv(file_path)
    # Filter rows where ISP contains the specified keyword (partial match, ignoring case)
    filtered_df = df[df['Org'].str.contains(isp, case=False, na=False)]
    # Also keep rows where Source IP is in the Destination IP
    video_ip_rows = df[df['Destination'].isin(filtered_df['Source'])]
    final_df = pd.concat([filtered_df, video_ip_rows]).drop_duplicates()
    # Save the filtered CSV
    save_path = file_path.replace('.csv', '_filtered.csv')
    filtered_df.to_csv(save_path, index=False)
    print(f"Filtered Video Streaming file saved to {save_path}")

# Function to walk through directories and filter files
def filter_files():
    # Process Online Gaming files
    for root, _, files in os.walk(online_gaming_dir):
        for file in files:
            if file.endswith('.csv'):
                file_path = os.path.join(root, file)
                filter_online_gaming(file_path)

    # Process Video Streaming files
    for folder, isp in video_folders_isps.items():
        folder_path = os.path.join(video_streaming_dir, folder)
        for root, _, files in os.walk(folder_path):
            for file in files:
                if file.endswith('.csv'):
                    file_path = os.path.join(root, file)
                    filter_video_streaming(file_path, isp)

# Run the filtering
filter_files()
```
